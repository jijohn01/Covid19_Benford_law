```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = pd.read_csv("./csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
target_country = 'global'
data
```

```python
if target_country=='global':
    pass
else:
    data= data[data['Country/Region']==target_country]
firstnums = np.asarray(data[data.columns[5:]].astype(str).apply(lambda a: [b[0] for b in a]),dtype=int).reshape(-1)
trueben =[ np.log10(1+1/i) for i in range(1,10)]
```

```python
p1=plt.hist(firstnums[firstnums!=0],density=True,rwidth=0.5,bins=range(1,11),align='left',color='lightgreen',label="Hist")
p2=plt.plot(range(1,10),trueben,marker='o',label='True')
plt.xticks(range(1,10))
plt.xlabel('1st Number')
plt.legend()
plt.title(str.upper(target_country)+" - Covid19 New Confirmed data - Benford's law result")
plt.show()
```
